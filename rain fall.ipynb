```python
data1=data.loc[(data.YEAR<=2010)]
```

```python
data2=data1.drop(['ANNUAL','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'],axis=1)
```

```python
data2['SUBDIVISION'].unique()
```

```python
data2=data2.loc[(data2.SUBDIVISION=="ARUNACHAL PRADESH")]
```

```python
data2
```

```python
list(data2)
```

```python
data2['ANNUAL RAIN']= data2.apply(lambda row: row.JAN + row.FEB,axis=1)
```

```python
data2
```

```python
cor=data2.corr()
cor
```

```python
sns.heatmap(cor,vmax=1,vmin=-1,annot=True,linewidths=.5,cmap='bwr')
```

```python
data2['SWM']= data2.apply(lambda row: row.JUN + row.JUL + row.AUG + row.SEP,axis=1)
```

```python
data2
```

```python
data2['NEM']= data2.apply(lambda row: row.OCT + row.NOV + row.DEC ,axis=1)
```
